<h2>Handling of Outliers in the dataset</h2>

```python
for i in columns:
    Q1=np.percentile(data[i],25,interpolation="midpoint")
    Q3=np.percentile(data[i],75,interpolation="midpoint")
    IQR=Q3-Q1
    UL=Q3+(1.5*IQR)
    LL=Q1-(1.5*IQR)
    data[i]=np.where(data[i]<LL,LL,np.where(data[i]>UL,UL,data[i]))
```

```python
columns=data.columns
print("\n\t\tBoxplot of the numerical features\n")
for i in enumerate(columns):
    plt.Figure(figsize=(10,10))
    plt.subplot(5,3,i[0]+1)
    sns.boxplot(x=i[1], data=data)
    plt.tight_layout()
```

outliers are removed from the dataset

<h3>Feature Reduction using PCA method</h3>

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.decomposition import PCA

scaler=StandardScaler()
# scaled_data=scaler.fit_transform(data)
# scaled_data=pd.DataFrame(scaled_data,columns=data.columns)
# scaled_data
pca=PCA(0.99)
pipe=make_pipeline(scaler,pca)
# Transform the data here
data_df=pipe.fit_transform(data)
data_df.shape
```

12 dimensions only required to have 99% varience in the data

```python
data_df=pd.DataFrame(data_df)
data_df2=data_df.copy()
data_df.head()
```

<h2>Clustering Models</h2>
</br>
<h3>1. KMeans</h3>

```python
#First we need to fin doptimal value of cluster- we rae using elbow graph
# importing necessary libraries
```

```python
from sklearn.cluster import KMeans
```

```python
#Find the optimal no of cluster
WCSS=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i, init="k-means++",random_state=42)
    kmeans.fit(data_df)
    WCSS.append(kmeans.inertia_)
plt.plot(range(1,11),WCSS,"-o")
```

optimal value of cluster from graph is 3

```python
#apply KNN model for k=3
model_kmeans=KMeans(n_clusters=3, init="k-means++",random_state=42)
model_kmeans.fit(data_df)
labels=model_kmeans.predict(data_df)
print(labels)
```

```python
# find the silhouette_score
from sklearn.metrics import silhouette_score
model_kmeans
score = silhouette_score(data_df, model_kmeans.labels_, metric='euclidean')
print('Silhouetter Score: ',score)
```

```python
clusterCut=pd.Series(labels)
```

```python
#clustered Dataset
data_kmhc = pd.concat([data_df, clusterCut], axis=1)
data_kmhc.columns = ['PC1','PC2','PC3','PC4','PC5','PC6','PC7','PC8','PC9','PC10','PC11','PC12',"ClusterID"]
data_kmhc.head()
```

<h3>2. Agglomerative Clustering</h3>

```python
# import necessary libraries
from scipy.cluster.hierarchy import dendrogram,linkage
```

```python
#finding optimal no of clusters using dendrogram
linked=linkage(data_df2, method="ward")
dendrogram(linked,orientation="top",distance_sort="acsending",show_leaf_counts=True)
plt.title("Dendrogram")
```

From the above graph we can see that optimal no of clusters is equal to 3

```python
# create the model using agglomerative clustering, n=3
#import necessary libraries

from sklearn.cluster import AgglomerativeClustering
cluster=AgglomerativeClustering(n_clusters=3,affinity="euclidean",linkage="ward")
labels_ag=cluster.fit_predict(data_df2)
labels_ag
```

```python
# find the silhouette_score
score = silhouette_score(data_df2, labels_ag, metric='euclidean')
print('Silhouetter Score: ',score)
```

```python
clusterCut=pd.Series(labels_ag)
```

```python
#clustered Dataset
data_kmhc = pd.concat([data_df2, clusterCut], axis=1)
data_kmhc.columns = ['PC1','PC2','PC3','PC4','PC5','PC6','PC7','PC8','PC9','PC10','PC11','PC12',"ClusterID"]
data_kmhc.head()
```
